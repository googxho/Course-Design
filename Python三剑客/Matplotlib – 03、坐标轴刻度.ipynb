### 1、设置坐标轴上下限

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

plt.plot(x,np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)

plt.show()
```

### 坐标轴也可以设置为倒序

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.xlim(11, -1)
plt.ylim(1.5, -1.5)

plt.show()
```

### 2、自定义刻度

```python

plt.plot(np.linspace(-45, 1, 1000))
plt.xticks([0,250,500,750,1000],rotation=30, fontsize='large')
plt.yticks([-45,-35,-25,-15,0],rotation=30, fontsize='small')
```
